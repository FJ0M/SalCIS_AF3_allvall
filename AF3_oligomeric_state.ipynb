```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.clf()
sns.set_context('paper')
sns.set_style('ticks')
plt.figure(0, (4, 4), dpi=300)

sns.pointplot(data=df, x='Oligomeric State', y='ipTM', legend=None,linestyle="--", errorbar=None,label='ipTM',
    marker="_", markersize=20, markeredgewidth=3,alpha=0.7)
sns.stripplot(
    data=df, x="Oligomeric State", y="ipTM", 
    dodge=True, alpha=.7, legend=False,
)

sns.pointplot(data=df, x='Oligomeric State', y='pTM', legend=None,linestyle="--", errorbar=None,label='pTM',
    marker="_", markersize=20, markeredgewidth=3,alpha=0.7)
sns.stripplot(
    data=df, x="Oligomeric State", y="pTM", 
    dodge=True, alpha=.7, legend=False,
)
plt.legend()
plt.ylim(0,1)
plt.ylabel('AlphaFold3 Model pTM or ipTM score\n(higher is better)')
plt.tight_layout()
plt.savefig('SalCIS13_oligomeric_state.png',transparent=True,dpi=600)
plt.show()
```
